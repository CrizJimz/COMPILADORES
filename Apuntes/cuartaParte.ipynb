El analizador sintáctico recibe el código fuente en la forma de tokens proveniente del analizador léxico y realiza el análisis sintáctico, que determina la estructura del programa. Esto es semejante a realizar el análisis gramatical sobre una frase en un lenguaje natural. El análisis sintáctico determina los elementos estructurales del programa y sus relaciones.
Los resultados del análisis sintáctico por lo regular se representan como un
árbol de análisis gramatical o un árbol sintáctico.

La función principal del analizador sintáctico (parser) es comprobar que los tokens que le suministra el analizador léxico van ordenados según la especificación de la gramática del lenguaje a compilar.
Y si no es as, dar los mensajes de error adecuados, pero continuar funcionando sin detenerse hasta que se llegue al final del fichero de entrada.
En pocas palabras guía todo el proceso, de la compilación, es debido a que va solicitando al analizador léxico los tokens y al mismo tiempo va dirigiendo el proceso de análisis semántico y generación de código intermedio. Por lo que muchas veces se lama al proceso de análisis semántico y generación de código intermedio, tradución dirigida por la sintaxis.

### Gramáticas

Gramáticas
Para lograr que un analizador sintáctico funcione, nosotros debemos especificar el lenguaje que debe poder leer. Para especificar este lenguaje, debemos representarlo con unas reglas únicas y bien formadas de manera que pueda funcionar de una manera bien definida, es decir indicar un lenguaje formal y a estas reglas se les llama gramática,
Una gramática independiente del contexto se define por tuplas de 4 elementos
que son:

•Terminales
•No Terminales
•Axioma Inicial
•Reglas de producción

G=(SUMAT, SUMAN,S,P)

### Gramáticas bien formadas

Gramáticas bien formadas
Cuando se tiene una gramática diseñada por expertos se espera que no tenga simbolos inútiles o reglas innecesarias, para evitar todo esto se necesita que la gramática bien formada, lo cual debemos hacer los siguiente:

• Quitar reglas Innecesarias
• Quitar simbolos inaccesibles
• Quitar simbolos superfluos
• No tener reglas no generativas
• No tener reglas de redenominación

Diremos que una gramática que cumple reglas innecesarias, inaccesibles y superfluos es una gramática limpia.

### Reglas innecesarias

Estas son las fáciles de identificar dado que se tiene reglas del tipo A=A, por lo que puede eliminar.

Ejemplo

G=({a,b,c},{S,A},S{S=aAb,})

### Simbolos inaccesbiles

Símbolos Inaccesibles
Este es el caso de que uno de los símbolos terminales y no terminales se puedan alcanzar, para evitar simbolos inacesibles podemos seguir los siguientes pasos:

1.- Construir una tabla con el número de filas igual a las reglas de producción, y columnas como símbolos ErUźn, se etiqueta cada columna con el símbolo terminal o no terminal.

2- Se marca un punto en la celda de la tabla donde esta el axioma

3.: Se señala con un punto las celdas de la parte derecha donde se marco el axioma; donde se contenga símbolos terminales y no terminales que correspondan a las regla de producción del axioma.

4: Para cada no terminal del paso 3, se marca (salvo el axioma) la celda, se vuelve a revisar todas las producciones marcando en la parte derecha de la celda antes marcada donde se contenga símbolos terminales y no terminales que correspondan a las regla de producción.

5, Se repite el paso 4, hasta que no quede ningún simbolo no terminal marcado por analizar, cuando se termina las columnas sin marcar son inaccesibles.

### Símbolos superfluos

Símbolos Superfluos
Es aquel donde un símbolo no terminal no permite generar desde él al menos una cadena vacía o que contenga solo símbolos terminales. Para no tener simbolos superfluos hacemos lo siguiente:

1.- Se construye una tabla que contenga el mismo número de filas que producciones, y se creara una columna que será la primera iteración, el número de iteraciones (columnas) crecerá dependiendo de las iteraciones necesarias.

2: En la primera Iteración marcaremos solo aquellas producciones que contengan en la parte derecha solo no terminales, cadenas de terminales o la cadena vacía A.

3.- Se crea una nueva iteración (columna), y se marcaran solo aquellas producciones que tengan en su parte derecha solo no ferminales del lado izquierdo de las producciones marcadas en la iteración (columna) anterior.

4. Repetir el paso 3, hasta que ya no se puedan marcar nuevas producciones, y las producciones que se quedan sin marcar son las superfluos

### Reglas no generativas

El formato de reglas de una gramática independiente del contexto indica que no deben existir reglas de reescritura compresoras, es decir producciones con lado derecho de menor longitud que el lado izquierdo.

Tomando en cuenta la regla A.=lambda), y que A no es el axioma de la gramática, es una regla compresora llamada regla no generativa. Si el axioma tiene en su producción la cadena vacía es decir :=A esta se permitirá como excepción y la llamaremos regla lambda.

Para quitar la regla no generativa hacemos los siguiente:

1-Para cada producción X:-aAB que se contenga el no terminal A en el lado derecho, agregar la regla de reescritura X:=aß, esto es porque se reemplaza A por lambda.

2-Eliminamos la producción A.=lambda.
