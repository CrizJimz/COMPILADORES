### Tratamiento de errores del Analizador Léxico

Simbolos no permitidos: Todo símbolo que no pertenezca al lenguaje que estemos analizando.
Identificador mal construido: Son aquellas palabras permitidas en el lenguaje que se ocupan como identificadores, por ejemplo un identificador que empiece con una letra y no con un número, y que tenga un tamaño determinado.
Valores numéricos mal construidos: Por ejemplo 12,34 en lugar de 12.34.

Constantes literales creadas erróneamente: Si se crea una cadena por ejemplo
- "México siempre fiel, esta mal construida porque falta el cierre de las dobles comillas.
- Podemos nosotros ignorar estos errores y tratarlos de la siguiente manera:

• Se ignoran los caracteres no validos hasta a completar el token.

• Insertar caracteres necesarios para a completar el token.

• Reemplazar el carácter incorrecto con el correcto para el token.

• Quitar caracteres inválidos.

Una de las partes importantes de un analizador léxico es poder reconocer entre una palabra reservada del lenguaje y un identificador, para lograr este paso tenemos dos posibles opciones:

•Crear una tabla con todas las palabras reservadas y consultarla cuando se crea un lexema que corresponda a un identificador.

•Implementar cada palabra reservada en el DT (se vera mas adelante), para poder distinguir entre los identificadores (esto complicaría el proceso de creación del programa del analizador léxico).

Depende de que tan grande sea el número de palabras reservadas se desean ocupar en el lenguaje, cualquiera de las dos opciones se tiene que consultar cuando se logre obtener un lexema para distinguir entre una palabra reservada o un identificador.

### Como crear un Analizador Léxico

Nuestro primer paso para crear un analizador léxico, es tratar de representarlo a través de una maquina de estados o que es lo mismo un diagrama de transiciones (DT). Sabemos que un existe una equivalencia con un Autómata
Finito Determinista (AFD) y el diagrama de transiciones.

Solo cabe mencionar que un AFD sólo dice si la secuencia de caracteres pertenece al lenguaje o no, y que nosotros necesitamos saber mediante el DT poder leer una secuencia completa hasta encontrar un símbolo (por ejemplo el espacio en blanco) que nos a complete un token. Una vez encontrado el token lo debemos regresar y estar listos para poder leer el siguiente token.

Tenemos que crear estados terminales y si se encuentra un lexema que no pertenezca a ninguna secuencia especificada se debe ir a un estado de error
para reportar.

### Autómata Finito Determinista

Un Autómata Finitio Determinista M es una quíntupla M = (Q,Z,P,q0,F) donde:

• Q es un conjunto finito de estados.

• Z es un alfabeto de entrada.

• P: Q x Z → Q es la función de transición que determina, para cada par formado por un estado y un símbolo de entrada, el estado siguiente al que pasa el autómata.

• q0 es el estado inicial.

• F Q es el conjunto de estados finales o de aceptación.

### Diferencia entre AFD y AFN

La diferencia entre AFD y AFN esta en la definición de la función de transición, dado que un AFN ante el símbolo de entrada puede parar a varios estados posibles, incluyendo a un único estado o a ninguno, esta nos dice que un AFN nos deja varias alternativas posibles o pude dejar ninguna.

### Expresiones regulares

Es una forma abreviada de representar cadenas de caracteres que se ajusten a un determinado patrón. Estas se definen sobre un alfabeto 2 y será una cadena formada por caracteres de dicho alfabeto. Las expresiones regulares básicas son:
1. El símbolo Ø (conjuto vacio) expresión regular L(Ø) = ( )
2. El símbolo 1 (palabra vacía) expresión regular L(2) = (4)
3. Cualquier a a e 2 es una expresión regular L(a) = {a)

### Operaciones con operaciones básicas

1. Concatenación: Ocupamos el metacaracter (.), si r y s son expresiones regulares entonces L(r.s)=L(r).L(s), se puede
omitir (.).
2. Unión: Se ocupa el metacaracter (|). Si r y s son expresiones
regulares entonces L(r/s) =L(r)|L(s).
3. Cierre o clausura: Se tiene el metacaracter * L(r*)=L(r)*.
