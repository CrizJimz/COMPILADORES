# ¿Que es un compilador?

Es un programa que convierte el programa escrito en un lenguaje (lenguaje fuente) a otro lenguaje (lenguaje destino).

### Diferencias entre compilador e interprete

El compilador traduce el programa completo y nos da un resultado final.

Mientras que el interprete ejecuta instrucción por instrucción y nos permite modificarlas durante la ejecución.

Pero los compiladoores corren más rápido los programas.

### Estructura de un compilador

La primer parte es de analisis y síntesis. (frontend y banckend).

### Analizador léxico

Esta fase del compilador efectúa la lectura real del programa fuente, el cual generalmente está en la forma de un flujo de caracteres. El rastreador realiza lo que se conoce como análisis léxico: recolecta secuencias de caracteres en unidades significativas denominadas componentes léxico o tokens, las cuales son como las palabras de un lenguaje natural, como el inglés o español. De este modo, se puede imaginar que un rastreador realiza una función similar al deletreo.

### Componentes léxicos o tokens

Los tipos de tokens que podemos tener son:

• Palabras reservadas: if, do, while, else

• Identificadores: nombres de variables, subprogramas, etiquetas

• Literales: cualquier carácter entre "* literales

• Números: en un sentido numérico (2,-45.34)

• Operadores y símbolos operacionales: +,*,<>=

### Vocabulario para un Analizador Léxico

Componente Léxico o token: Van a hacer los elementos del lenguaje, en donde las demás componentes del compilador son los simbolos terminales de la gramática. Un ejemplo de los componentes son las palabras reservadas, signos de puntuación, identificadores, valores
numéricos etc.

Lexema: Es la secuencia de caracteres del código fuente, que se mapea con el patrón de un token, es decir cuando se analiza el código fuente y se encuentra una cadena de caracteres que se mapea con un token, se dice que la cadena es un token.

Tabla de tokens: Estructura de dato que se almacena todos los tokens del lenguaje.

Tabla de simbolos: Sera necesario identificar los identificadores y valores numéricos que empleara el programa que queremos compilar.

### Fase del Analizador Léxico

En esta fase, la cadena de caracteres que constituye el programa fuente se lee de izquierda a derecha y se agrupa en componentes léxicos conocidos como lexemas, y cada lexema produce una salida que la llamaremos tokens, que son secuencias de caracteres que tienen un significado atómico; además el analizador léxico trabaja con la tabla de símbolos introduciendo en ésta los nombres de las variables.

Para cada token que se produce, se tendrá la siguiente salida:

<nombre del token, valor-atributo>

### Tratamiento de errores del Analizador Léxico

Simbolos no permitidos: Todo símbolo que no pertenezca al lenguaje que estemos analizando.
Identificador mal construido: Son aquellas palabras permitidas en el lenguaje que se ocupan como identificadores, por ejemplo un identificador que empiece con una letra y no con un número, y que tenga un tamaño determinado.
Valores numéricos mal construidos: Por ejemplo 12,34 en lugar de 12.34.

Constantes literales creadas erróneamente: Si se crea una cadena por ejemplo
- "México siempre fiel, esta mal construida porque falta el cierre de las dobles comillas.
Podemos nosotros ignorar estos errores y tratarlos de la siguiente manera:
• Se ignoran los caracteres no validos hasta a completar el token.
• Insertar caracteres necesarios para a completar el token.
• Reemplazar el carácter incorrecto con el correcto para el token.
• Quitar caracteres inválidos.
